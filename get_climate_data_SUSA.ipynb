```python
import requests
from datetime import datetime, timedelta


susa_stations = {'FREJUS':'PIE-001022-901-1990-11-09', 'BORGONE':'PIE-001032-900-1991-01-10', 'SALBELTRAND':'PIE-001232-900-1990-11-15', 'SUSA':'PIE-001270-900-1990-11-14', 'AVIGLIANA':'PIE-001013-900-1991-01-09'}
params = ['T', 'H', 'R', 'V', 'I']
```

```python
def get_data(station, param):
    url = "https://utility.arpa.piemonte.it/"
    url += "meteoidro/dati_giornalieri_meteo/"
    params = {
        "codice": station,
        "data_inizio": (datetime.today() - timedelta(days=3650)).strftime("%Y-%m-%d"),
        "data_fine": datetime.today().strftime("%Y-%m-%d"),
        "param": param
    }
    return requests.get(url, params=params).json()
```

```python
data = []
for city, station_code in susa_stations.items():
    for param in params:
        data.append(get_data(station_code, param))
```
