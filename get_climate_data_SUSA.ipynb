```python
import requests
from datetime import datetime, timedelta
import pandas as pd
```

```python
susa_stations = {'PRERICHARD':'PIE-001022-900', 'BORGONE':'PIE-001032-900', 'SALBELTRAND':'PIE-001232-900', 'AVIGLIANA':'PIE-001013-900'}
```

```python
def get_data(station, start_date, end_date):
    url = "https://utility.arpa.piemonte.it/meteoidro/dati_giornalieri_meteo/"
    params = {
        "fk_id_punto_misura_meteo": station,
        "data_min": start_date,
        "data_max": end_date
    }
    return requests.get(url, params=params).json()
```

```python
dfs_city = {}
for city, station_code in susa_stations.items():
    dfs = []   
    for year in range(1990, 2026):
        start_date = f"{year}-01-01"
        end_date = f"{year}-12-31"
        data = get_data(station_code, start_date, end_date)
        res = data.get('results')
        if len(res) > 0:
            df = pd.DataFrame(res)
            df = df.drop(['url', 'fk_id_punto_misura_meteo'], axis = 1)
            df = df.set_index('data')
            dfs.append(df)
        else:
            print(city, year)
    df_city = pd.concat(dfs, axis = 0)
    dfs_city[city] = df_city
    df_city.to_csv('data/' + city + '.csv', index_label=True)
```
